```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical
```

```python
df = pd.read_csv("music_genre.csv")
df
```

```python
le_key = LabelEncoder()
le_mode = LabelEncoder()
df["key"] = le_key.fit_transform(df["key"])
df["mode"] = le_mode.fit_transform(df["mode"])
df["tempo"] = pd.to_numeric(df["tempo"], errors="coerce")
df.dropna(subset=["tempo"], inplace=True)
df
```

```python
features = [
    "popularity",
    "acousticness",
    "danceability",
    "duration_ms",
    "energy",
    "instrumentalness",
    "key",
    "liveness",
    "loudness",
    "mode",
    "speechiness",
    "tempo",
    "valence",
]

target = "music_genre"

X = df[features]
y = df[target]
```

```python
le_genre = LabelEncoder()
y_encoded = le_genre.fit_transform(y)
y_categorical = to_categorical(y_encoded)
```

```python
X_temp, X_real_test, y_temp, y_real_test = train_test_split(
    X, y_categorical, test_size=0.2, random_state=42
)

X_train, X_val, y_train, y_val = train_test_split(
    X_temp, y_temp, test_size=0.25, random_state=42
)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
X_real_test_scaled = scaler.transform(X_real_test)
```

```python
model = Sequential()
model.add(Dense(128, input_dim=X_train_scaled.shape[1], activation="relu"))
model.add(Dense(64, activation="relu"))
model.add(Dense(32, activation="relu"))
model.add(Dense(y_categorical.shape[1], activation="sigmoid"))
```

```python
model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
history = model.fit(
    X_train_scaled,
    y_train,
    epochs=100,
    validation_data=(X_val_scaled, y_val),
)
```

```python
real_test_loss, real_test_accuracy = model.evaluate(X_real_test_scaled, y_real_test)
print(f"Real Test Accuracy: {real_test_accuracy:.3f}")
```
